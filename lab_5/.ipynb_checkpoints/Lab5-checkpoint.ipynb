```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Read data

```python
f = open("data.csv")
res= []
for i in f:
    res.append(i.strip().split(","))

tr = TransactionEncoder()
tr_ar = tr.fit(res).transform(res)
data = pd.DataFrame(tr_ar, columns=tr.columns_)
data
```

# Apriori
## We will be using min support = 0.01
Other values will be investigated later

```python
%%timeit -n 50
apriori_DF = apriori(data, use_colnames=True, min_support=0.01)
```

```python
apriori_DF = apriori(data, use_colnames=True, min_support=0.01)
```

```python
apriori_DF
```

```python
rules = association_rules(apriori_DF, metric='lift', min_threshold=1).sort_values(by=['confidence'], ascending=False)
rules
```

```python
print("Confidence_min = ", rules.confidence.min())
```

## investigating values of min_support in [0.01, 0.06]

```python
for i in np.linspace(0.01, 0.06, 6):
    apriori_DFi = apriori(data, use_colnames=True, min_support=i)
    rulesi = association_rules(apriori_DFi, metric='lift', min_threshold=1).sort_values(by=['confidence'], ascending=False)
    print(f"\n\n\ni={i}")
    display(apriori_DFi)
    display(rulesi)
    print("Confidence_min = ", rulesi.confidence.min())
    
```

Звідси бачимо, що найкраще значення min_support = 0.05, бо він має найкращий параметр confidence, після значення 0.06 усі отримані набори - одноелементні, тому правила не можуть бути створені, плак

# fp growth

```python
%%timeit -n 50
fpgrowth_DF = fpgrowth(data, use_colnames=True, min_support=0.01)
```

```python
fpgrowth_DF = fpgrowth(data, use_colnames=True, min_support=0.01)
```

```python
fpgrowth_DF
```

```python
rules = association_rules(fpgrowth_DF, metric='lift', min_threshold=1).sort_values(by=['confidence'], ascending=False)
rules
```

```python
print("Confidence_min = ", rules.confidence.min())
```

## investigating values of min_support in [0.01, 0.06]

```python
X = np.linspace(0.01, 0.06, 6)
Y1 = []
Y2 = []
for i in np.linspace(0.01, 0.06, 6):
    fpgrowth_DFi = fpgrowth(data, use_colnames=True, min_support=i)
    rulesi = association_rules(fpgrowth_DFi, metric='lift', min_threshold=1).sort_values(by=['confidence'], ascending=False)
    print(f"\n\n\ni={i}")
    display(fpgrowth_DFi)
    display(rulesi)
    Y1.append(rulesi.confidence.min())
    Y2.append(len(rulesi))
    print("Confidence_min = ", rulesi.confidence.min())
    
```

```python
plt.plot(X, Y1)
```

```python
plt.plot(X, Y2)
```

# Висновок

Бачимо, що методи побудови ФП-Росту та АПРІОРІ дають однаковий набір правил, причому він найкращий (тобто з найбільшим параметром min_confidence) при значенні **min_support=0.05**
**min_confidence=0.213**

Але при цьому алгоритм ФП-Росту працює швидше, аніж АПРІОРІ, тому краще використовувати його.
