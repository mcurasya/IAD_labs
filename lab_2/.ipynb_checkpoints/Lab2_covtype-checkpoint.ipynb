```python
from sklearn.datasets import fetch_covtype
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, recall_score, precision_score, auc, confusion_matrix, precision_recall_curve, roc_curve, PrecisionRecallDisplay, RocCurveDisplay, classification_report
from sklearn.model_selection import GridSearchCV


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, Y = fetch_covtype(return_X_y=True, shuffle=True)
X, _, Y, _ = train_test_split(X, Y, train_size=0.0005)
#this is terrible but my pc is too slow, and I am still waiting
np.unique(Y)
X.shape
```

Центруємо наші дані

```python
X = (X - np.mean(X, axis=0))
X_7 = X[Y==7]
X_6 = X[Y==6]
X_5 = X[Y==5]
X_4 = X[Y==4]
X_3 = X[Y==3]
X_2 = X[Y==2]
X_1 = X[Y==1]
X_0 = X[Y==0]
plt.scatter(X_7[:, 0], X_7[:, 1], label="X=7")
plt.scatter(X_6[:, 0], X_6[:, 1], label="X=6")
plt.scatter(X_5[:, 0], X_5[:, 1], label="X=5")
plt.scatter(X_4[:, 0], X_4[:, 1], label="X=4")
plt.scatter(X_3[:, 0], X_3[:, 1], label="X=3")
plt.scatter(X_2[:, 0], X_2[:, 1], label="X=2")
plt.scatter(X_1[:, 0], X_1[:, 1], label="X=1")
plt.scatter(X_0[:, 0], X_0[:, 1], label="X=0")
plt.legend()
```

Навчимо LinearSVC на центрованій вибірці

Все погано, але тому, що довго

Виконаємо 5fold крос-валідацію

```python
clf = LinearSVC(C=1)
cross_val_score(clf, X, Y)
```

Порахуємо метрики

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=0)
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
Y_pred
```

```python
recall_score(Y_test, Y_pred, average=None)
```

```python
precision_score(Y_test, Y_pred, average=None)
```

```python
print(classification_report(Y_test, Y_pred))
```

```python
confusion_matrix(Y_test, Y_pred)
```

Побудуємо для тієї самої центрованої вибірки SVC зі значеннями ядра linear та poly

```python
clf_lin = SVC(kernel="linear", C=1)
clf_poly = SVC(kernel="poly", C=10)
cross_val_score(clf_poly, X, Y)
```

```python
clf_poly.fit(X_train, Y_train)
Y_pred = clf_poly.predict(X_test)
```

Виконаємо grid-search для параметрів C, degree, probability 

```python
params = {'probability':(True, False), 'C':[1, 10, 100, 1000], 'degree': [1,3,5,7]}
clf_searched = SVC(kernel='poly')
gs_clf = GridSearchCV(clf_searched, params)
gs_clf.fit(X_train, Y_train)
```

```python
gs_clf.best_params_
```

```python
Y_best_pred = gs_clf.predict(X_test)
```

```python
confusion_matrix(Y_test, Y_best_pred)
```

```python
print(classification_report(Y_test, Y_best_pred))
```
